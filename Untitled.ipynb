**–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:**
- –§–∞–π–ª Excel —Å –º—É–ª—å—Ç–∏—É—Ä–æ–≤–Ω–µ–≤—ã–º–∏ –∑–∞–≥–æ–ª–æ–≤–∫–∞–º–∏
- –°—Ç—Ä—É–∫—Ç—É—Ä–∞: –ì–æ–¥ ‚Üí –ú–µ—Å—è—Ü ‚Üí –¢–æ–≤–∞—Ä + –¢–∏–ø ‚Üí –î–∞–Ω–Ω—ã–µ –ø–æ 20 –º–∞–≥–∞–∑–∏–Ω–∞–º

**–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:**
- –ü–ª–æ—Å–∫–∞—è —Ç–∞–±–ª–∏—Ü–∞ —Å –∫–æ–ª–æ–Ω–∫–∞–º–∏: –ì–æ–¥, –ú–µ—Å—è—Ü, –¢–æ–≤–∞—Ä, –¢–∏–ø, –ú–∞–≥–∞–∑–∏–Ω, –ß–∏—Å–ª–æ —á–µ–∫–æ–≤, –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤ —á–µ–∫–µ, –°—É–º–º–∞ –≤ —á–µ–∫–µ, –ù–∞—Ü–µ–Ω–∫–∞ –ø—Ä–æ–¥–∞–∂–∏ –≤ —á–µ–∫–µ

```python
# –ò–º–ø–æ—Ä—Ç –±–∏–±–ª–∏–æ—Ç–µ–∫
import pandas as pd
import numpy as np
from pathlib import Path
import warnings
warnings.filterwarnings('ignore')

print("‚úì –ë–∏–±–ª–∏–æ—Ç–µ–∫–∏ –∑–∞–≥—Ä—É–∂–µ–Ω—ã")
```

## üìÇ –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –ø—É—Ç–µ–π –∫ —Ñ–∞–π–ª–∞–º

```python
# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã —Ñ–∞–π–ª–æ–≤
INPUT_FILE = "raw_data.xlsx"  # –ò—Å—Ö–æ–¥–Ω—ã–π —Ñ–∞–π–ª –≤ —à–∏—Ä–æ–∫–æ–º —Ñ–æ—Ä–º–∞—Ç–µ
OUTPUT_FILE = "final_flat_clean.xlsx"  # –†–µ–∑—É–ª—å—Ç–∞—Ç –≤ –ø–ª–æ—Å–∫–æ–º —Ñ–æ—Ä–º–∞—Ç–µ
SHEET_NAME = 0  # –ü–µ—Ä–≤—ã–π –ª–∏—Å—Ç

# –ü—Ä–æ–≤–µ—Ä—è–µ–º –Ω–∞–ª–∏—á–∏–µ —Ñ–∞–π–ª–∞
if Path(INPUT_FILE).exists():
    print(f"‚úì –ù–∞–π–¥–µ–Ω —Ñ–∞–π–ª: {INPUT_FILE}")
else:
    print(f"‚ö†Ô∏è  –§–∞–π–ª –Ω–µ –Ω–∞–π–¥–µ–Ω: {INPUT_FILE}")
    print("   –°–æ–∑–¥–∞–π—Ç–µ —Ñ–∞–π–ª –∏–ª–∏ –∏–∑–º–µ–Ω–∏—Ç–µ –ø—É—Ç—å –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π INPUT_FILE")
```

## üìä –ß—Ç–µ–Ω–∏–µ –∏—Å—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö

```python
# –ß–∏—Ç–∞–µ–º —Ç–∞–±–ª–∏—Ü—É —Å –¥–≤—É–º—è —Å—Ç—Ä–æ–∫–∞–º–∏ –∑–∞–≥–æ–ª–æ–≤–∫–∞
df_wide = pd.read_excel(INPUT_FILE, sheet_name=SHEET_NAME, header=[0, 1])

print(f"‚úì –î–∞–Ω–Ω—ã–µ –∑–∞–≥—Ä—É–∂–µ–Ω—ã")
print(f"  –†–∞–∑–º–µ—Ä —Ç–∞–±–ª–∏—Ü—ã: {df_wide.shape[0]} —Å—Ç—Ä–æ–∫ √ó {df_wide.shape[1]} –∫–æ–ª–æ–Ω–æ–∫")
print(f"\n–ü—Ä–µ–≤—å—é –ø–µ—Ä–≤—ã—Ö —Å—Ç—Ä–æ–∫:")
df_wide.head()
```

## üîç –ü–∞—Ä—Å–∏–Ω–≥ –∑–∞–≥–æ–ª–æ–≤–∫–æ–≤

```python
def parse_headers(df):
    """
    –ò–∑–≤–ª–µ–∫–∞–µ—Ç —Å–ø–∏—Å–æ–∫ –º–∞–≥–∞–∑–∏–Ω–æ–≤ –∏ –º–µ—Ç—Ä–∏–∫ –∏–∑ –º—É–ª—å—Ç–∏—É—Ä–æ–≤–Ω–µ–≤—ã—Ö –∑–∞–≥–æ–ª–æ–≤–∫–æ–≤.
    """
    stores = []
    metrics = []
    
    for col in df.columns:
        store_name = col[0] if not pd.isna(col[0]) and 'Unnamed' not in str(col[0]) else None
        metric_name = col[1] if not pd.isna(col[1]) and 'Unnamed' not in str(col[1]) else None
        
        if store_name and store_name not in stores:
            stores.append(store_name)
        if metric_name and metric_name not in metrics:
            metrics.append(metric_name)
    
    return stores, metrics

# –ü–∞—Ä—Å–∏–º –∑–∞–≥–æ–ª–æ–≤–∫–∏
stores, metrics = parse_headers(df_wide)

print(f"‚úì –ù–∞–π–¥–µ–Ω–æ –º–∞–≥–∞–∑–∏–Ω–æ–≤: {len(stores)}")
print(f"  –°–ø–∏—Å–æ–∫: {stores[:5]}..." if len(stores) > 5 else f"  –°–ø–∏—Å–æ–∫: {stores}")
print(f"\n‚úì –ù–∞–π–¥–µ–Ω–æ –º–µ—Ç—Ä–∏–∫: {len(metrics)}")
print(f"  –°–ø–∏—Å–æ–∫: {metrics}")
```

## üîÑ –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –≤ –ø–ª–æ—Å–∫–∏–π —Ñ–æ—Ä–º–∞—Ç

```python
def transform_to_flat(df):
    """
    –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ—Ç —à–∏—Ä–æ–∫—É—é —Ç–∞–±–ª–∏—Ü—É –≤ –ø–ª–æ—Å–∫—É—é.
    """
    # –ü–µ—Ä–≤–∞—è –∫–æ–ª–æ–Ω–∫–∞ - —ç—Ç–æ –∏–Ω–¥–µ–∫—Å (–ù–∞–∑–≤–∞–Ω–∏–µ —Å—Ç—Ä–æ–∫ / –¢–∏–ø)
    index_col = df.columns[0]
    df_index = df[index_col].copy()
    
    # –°–ø–∏—Å–æ–∫ –¥–ª—è –Ω–∞–∫–æ–ø–ª–µ–Ω–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    result_rows = []
    
    # –¢–µ–∫—É—â–∏–π –≥–æ–¥ –∏ –º–µ—Å—è—Ü
    current_year = None
    current_month = None
    
    # –ü—Ä–æ—Ö–æ–¥–∏–º –ø–æ –≤—Å–µ–º —Å—Ç—Ä–æ–∫–∞–º
    for idx, row in df.iterrows():
        index_value = df_index.iloc[idx]
        
        # –ü—Ä–æ–ø—É—Å–∫–∞–µ–º NaN
        if pd.isna(index_value):
            continue
        
        index_str = str(index_value).strip()
        
        # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —ç—Ç–æ –≥–æ–¥?
        if index_str.isdigit() and len(index_str) == 4:
            current_year = int(index_str)
            current_month = None
            continue
        
        # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —ç—Ç–æ –º–µ—Å—è—Ü?
        months_ru = [
            "–Ø–Ω–≤–∞—Ä—å", "–§–µ–≤—Ä–∞–ª—å", "–ú–∞—Ä—Ç", "–ê–ø—Ä–µ–ª—å", "–ú–∞–π", "–ò—é–Ω—å",
            "–ò—é–ª—å", "–ê–≤–≥—É—Å—Ç", "–°–µ–Ω—Ç—è–±—Ä—å", "–û–∫—Ç—è–±—Ä—å", "–ù–æ—è–±—Ä—å", "–î–µ–∫–∞–±—Ä—å"
        ]
        if index_str in months_ru:
            current_month = index_str
            continue
        
        # –ò–Ω–∞—á–µ —ç—Ç–æ —Å—Ç—Ä–æ–∫–∞ —Å —Ç–æ–≤–∞—Ä–æ–º
        # –ü–∞—Ä—Å–∏–º —Ç–æ–≤–∞—Ä –∏ —Ç–∏–ø (—Ä–∞–∑–¥–µ–ª–µ–Ω—ã —Ç–∞–±—É–ª—è—Ü–∏–µ–π –∏–ª–∏ –Ω–µ—Å–∫–æ–ª—å–∫–∏–º–∏ –ø—Ä–æ–±–µ–ª–∞–º–∏)
        parts = index_str.split('\t')
        if len(parts) == 1:
            # –ü–æ–ø—Ä–æ–±—É–µ–º —Ä–∞–∑–¥–µ–ª–∏—Ç—å –ø–æ –¥–≤—É–º –∏–ª–∏ –±–æ–ª–µ–µ –ø—Ä–æ–±–µ–ª–∞–º
            parts = [p.strip() for p in index_str.split('  ') if p.strip()]
        
        if len(parts) >= 2:
            tovar = parts[0].strip()
            tip = parts[1].strip()
        else:
            # –ï—Å–ª–∏ –Ω–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å —Ä–∞—Å–ø–∞—Ä—Å–∏—Ç—å, –ø—Ä–æ–ø—É—Å–∫–∞–µ–º
            continue
        
        # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ –≥–æ–¥ –∏ –º–µ—Å—è—Ü –æ–ø—Ä–µ–¥–µ–ª–µ–Ω—ã
        if current_year is None or current_month is None:
            continue
        
        # –ò–∑–≤–ª–µ–∫–∞–µ–º –¥–∞–Ω–Ω—ã–µ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –º–∞–≥–∞–∑–∏–Ω–∞
        for store in stores:
            row_data = {
                '–ì–æ–¥': current_year,
                '–ú–µ—Å—è—Ü': current_month,
                '–¢–æ–≤–∞—Ä': tovar,
                '–¢–∏–ø': tip,
                '–ú–∞–≥–∞–∑–∏–Ω': store
            }
            
            # –ò–∑–≤–ª–µ–∫–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –º–µ—Ç—Ä–∏–∫
            for metric in metrics:
                try:
                    value = row[(store, metric)]
                    row_data[metric] = value if not pd.isna(value) else 0
                except KeyError:
                    row_data[metric] = 0
            
            result_rows.append(row_data)
    
    # –°–æ–∑–¥–∞—ë–º –∏—Ç–æ–≥–æ–≤—ã–π DataFrame
    df_flat = pd.DataFrame(result_rows)
    
    return df_flat

# –í—ã–ø–æ–ª–Ω—è–µ–º –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ
print("üîÑ –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö...")
df_flat = transform_to_flat(df_wide)

print(f"‚úì –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –∑–∞–≤–µ—Ä—à–µ–Ω–æ!")
print(f"  –†–∞–∑–º–µ—Ä –ø–ª–æ—Å–∫–æ–π —Ç–∞–±–ª–∏—Ü—ã: {df_flat.shape[0]:,} —Å—Ç—Ä–æ–∫ √ó {df_flat.shape[1]} –∫–æ–ª–æ–Ω–æ–∫")
```

## ‚úÖ –í–∞–ª–∏–¥–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤

```python
# –ü—Ä–æ–≤–µ—Ä—è–µ–º —Å—Ç—Ä—É–∫—Ç—É—Ä—É
print("üìä –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –ø–ª–æ—Å–∫–æ–π —Ç–∞–±–ª–∏—Ü–µ:")
print(f"\n{'='*60}")
print(f"–°—Ç—Ä–æ–∫: {len(df_flat):,}")
print(f"\n–ö–æ–ª–æ–Ω–∫–∏:")
for col in df_flat.columns:
    print(f"  ‚Ä¢ {col}")

print(f"\n{'='*60}")
print(f"–£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π:")
print(f"  ‚Ä¢ –ì–æ–¥–æ–≤: {df_flat['–ì–æ–¥'].nunique()}")
print(f"  ‚Ä¢ –ú–µ—Å—è—Ü–µ–≤: {df_flat['–ú–µ—Å—è—Ü'].nunique()}")
print(f"  ‚Ä¢ –¢–æ–≤–∞—Ä–æ–≤: {df_flat['–¢–æ–≤–∞—Ä'].nunique()}")
print(f"  ‚Ä¢ –¢–∏–ø–æ–≤ —Ç–æ–≤–∞—Ä–æ–≤: {df_flat['–¢–∏–ø'].nunique()}")
print(f"  ‚Ä¢ –ú–∞–≥–∞–∑–∏–Ω–æ–≤: {df_flat['–ú–∞–≥–∞–∑–∏–Ω'].nunique()}")

print(f"\n{'='*60}")
print(f"–ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è:")
missing = df_flat.isna().sum()
if missing.sum() == 0:
    print("  ‚úì –ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –Ω–µ—Ç")
else:
    print(missing[missing > 0])
```

## üíæ –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞

## üéØ –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ü–µ–ª–æ—Å—Ç–Ω–æ—Å—Ç–∏ –¥–∞–Ω–Ω—ã—Ö

```python
# –ü—Ä–æ–≤–µ—Ä—è–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–ø–∏—Å–µ–π
expected_years = df_flat['–ì–æ–¥'].nunique()
expected_months = df_flat['–ú–µ—Å—è—Ü'].nunique()
expected_products = df_flat['–¢–æ–≤–∞—Ä'].nunique()
expected_stores = df_flat['–ú–∞–≥–∞–∑–∏–Ω'].nunique()

# –û–∂–∏–¥–∞–µ–º–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫
expected_rows = expected_years * expected_months * expected_products * expected_stores

print(f"üìä –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ü–µ–ª–æ—Å—Ç–Ω–æ—Å—Ç–∏ –¥–∞–Ω–Ω—ã—Ö:")
print(f"\n–û–∂–∏–¥–∞–µ–º–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫:")
print(f"  {expected_years} –≥–æ–¥–æ–≤ √ó {expected_months} –º–µ—Å—è—Ü–µ–≤ √ó {expected_products} —Ç–æ–≤–∞—Ä–æ–≤ √ó {expected_stores} –º–∞–≥–∞–∑–∏–Ω–æ–≤ = {expected_rows:,} —Å—Ç—Ä–æ–∫")
print(f"\n–§–∞–∫—Ç–∏—á–µ—Å–∫–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫: {len(df_flat):,}")
print(f"\n–†–∞–∑–Ω–∏—Ü–∞: {abs(expected_rows - len(df_flat)):,} —Å—Ç—Ä–æ–∫")

if expected_rows == len(df_flat):
    print("\n‚úÖ –î–∞–Ω–Ω—ã–µ –ø–æ–ª–Ω—ã–µ, –ø—Ä–æ–ø—É—Å–∫–æ–≤ –Ω–µ—Ç!")
else:
    coverage = (len(df_flat) / expected_rows * 100) if expected_rows > 0 else 0
    print(f"\n‚ö†Ô∏è  –ü–æ–∫—Ä—ã—Ç–∏–µ –¥–∞–Ω–Ω—ã—Ö: {coverage:.1f}%")
```
